```python
from constraints import *
def print_cost_summary(assignment, domain, constraints_dict):
    skill = lack_required_skill(assignment, domain, constraints_dict['lack_required_skill'])
    many_workers = too_many_workers(assignment, domain, constraints_dict['too_many_workers'])
    few_workers = too_few_workers(assignment, domain, constraints_dict['too_few_workers'])
    many_tasks = too_many_tasks(assignment, domain, constraints_dict['too_many_tasks'])
    few_tasks = too_few_tasks(assignment, domain, constraints_dict['too_few_tasks'])
    print "SKILL", skill
    print "TOO MANY WORKERS", many_workers
    print "TOO FEW WORKERS", few_workers
    print "TOO MANY TASKS", many_tasks
    print "TOO FEW TASKS", few_tasks
    cost = skill + many_workers + many_tasks + few_workers + few_tasks
    print "TOTAL: ", cost
    return cost
```

```python
import sys
import worker
import task
import local
import csv

def run(algo, worker_csv, task_csv):
    # initialize workers
    with open(worker_csv, "rb") as workersFile:
        reader = csv.reader(workersFile)
        skillList = next(reader, None)[1:-1]
        workers = {}
        for row in reader:
            name = row[0]
            skillsTF = row[1:-1] # e.g. ['T', 'F', 'T', 'F']
            skills = [] # e.g. ['Computer', 'Sorting']
            for i in xrange(len(skillsTF)):
                if skillsTF[i] == 'T':
                    skills.append(skillList[i])
            wanted_tasks = row[-1]
            workers[name] = worker.Worker(name, skills = skills, wanted_tasks = wanted_tasks)
    
    # initialize tasks
    with open(task_csv, "rb") as tasksFile:
        reader = csv.reader(tasksFile)
        skillList = next(reader, None)[1:-1]
        tasks = {}
        for row in reader:
            taskNumber = row[0]
            skillsTF = row[1:-1] # e.g. ['T', 'F', 'T', 'F']
            skills = [] # e.g. ['Computer', 'Sorting']
            for i in xrange(len(skillsTF)):
                if skillsTF[i] == 'T':
                    skills.append(skillList[i])
            wanted_workers = row[-1]
            tasks[taskNumber] = task.Task(taskNumber, skill_reqs = skills, wanted_workers = wanted_workers)
    
    # constraints_dict
    constraints_dict = {
        'lack_required_skill' : 200,
        'too_many_workers' : 10,
        'too_few_workers' : 20,
        'too_many_tasks' : 10,
        'too_few_tasks' : 20
    }
    
    # get algorithm name argument
    solution = getattr(local, algo)((workers, tasks), constraints_dict)
    cost = print_cost_summary(solution[0], (workers, tasks), constraints_dict)
    print cost
    return (solution, cost, workers, tasks)
```

```python
def run_tests(worker_csv, task_csv):
    clock_runtimes = {}
    nodes_expanded = {}
    costs = {}
    for alg in algorithms:
        t = stopwatch.Timer()    
        sol, cost, workers, tasks = run(alg, worker_csv, task_csv)
        assignment, nodes = sol
        t.stop()
        clock_runtimes[alg] = t.elapsed
        nodes_expanded[alg] = nodes
        costs[alg] = cost
        print "Algo analyzed"
    return [clock_runtimes, nodes_expanded, costs, workers, tasks]
```

```python
def plot_bar(title, data, workers, tasks, c=1):
    fig = plt.figure(figsize=(15, 7))
    plt.title(title + " for algorithms on DataSet of " + str(workers) + " workers, and " + str(tasks) + " tasks")
    x = np.arange(len(algorithms))
    y = np.array(data.values())
    
    plt.bar(x, y, color=dark2_colors[c], align='center', width=.5)
    
    plt.xlabel("Algorithm")
    plt.ylabel(title)
    remove_border()
    ticks = [str(i)+ " (" + str(round(data[i],2)) + ")" for i in data.keys()]
    ticks = plt.xticks(x, ticks, 
                       rotation='-10', fontsize=10)
    limits = plt.xlim(-1, x[-1] + 1)
    plt.show()
```

```python
def plots(data):
    plot_bar("Clock Runtime", data[0], len(data[3]), len(data[4]), 0)
    plot_bar("Node Expansion Runtime", data[1], len(data[3]), len(data[4]), 1)
    plot_bar("Cost Effectiveness", data[2], len(data[3]), len(data[4]), 2)
```
